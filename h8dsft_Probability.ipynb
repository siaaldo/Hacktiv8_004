```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
head_tail = [0,0]

trials = 50000
trial = 0

while trial < trials :
    trial+=1
    randomToss = np.random.randint(2)
    head_tail[randomToss] = head_tail[randomToss] + 1
```

```python
plt.figure(figsize=(5,5))
plt.pie(head_tail, labels=['head', 'tail'])
plt.legend()
plt.show()
```
